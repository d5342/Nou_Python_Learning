```python
import matplotlib.pyplot as plt
X=range(1,20)
Y=[value*2 for value in X]
print('Values of X:')
print(*range(1,20))
print('Values of Y:')
print(Y)
plt.plot(X,Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.show()
```

```python
import matplotlib.pyplot as plt

# 
with open("mydata.txt") as f:
    data = f.read()

# 
data = [line for line in data.split('\n') if line.strip()]

#  x  y 
x = [int(row.split(' ')[0]) for row in data]  #  x
y = [int(row.split(' ')[1]) for row in data]  #  y

# 
print(x)
print(y)

# 
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 
data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03"],
    "Value": [10, 20, 15]
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])  # 
df.set_index("Date", inplace=True)      # 

# 
df.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]  # 

print(x_pos)

plt.bar(x_pos, popularity, color=(0.4, 0.6, 0.8, 1.0))
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\n' + 'Worldwide, Oct. 2025 compared to a year ago')  # 
plt.xticks(x_pos, x)  #  x_pos  x 

# 
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')  #  which
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')  #  which

plt.show()
```
