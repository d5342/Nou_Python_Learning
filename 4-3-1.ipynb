```python
import matplotlib.pyplot as plt

# 讀取檔案
with open("mydata.txt") as f:
    data = f.read()

# 按行分割，並過濾掉空行
data = [line for line in data.split('\n') if line.strip()]

# 提取 x 和 y 座標
x = [int(row.split(' ')[0]) for row in data]  # 第一欄作為 x
y = [int(row.split(' ')[1]) for row in data]  # 第二欄作為 y

# 印出檢查
print(x)
print(y)

# 繪製圖表
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Sample graph!')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 直接創建資料
data = {
    "Date": ["2023-01-01", "2023-01-02", "2023-01-03"],
    "Value": [10, 20, 15]
}
df = pd.DataFrame(data)
df["Date"] = pd.to_datetime(df["Date"])  # 轉換為日期格式
df.set_index("Date", inplace=True)      # 設置日期為索引

# 繪圖
df.plot()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]  # 修正拼寫和語法

print(x_pos)

plt.bar(x_pos, popularity, color=(0.4, 0.6, 0.8, 1.0))
plt.xlabel('Languages')
plt.ylabel('Popularity')
plt.title('Popularity of Programming Language\n' + 'Worldwide, Oct. 2025 compared to a year ago')  # 修正拼寫
plt.xticks(x_pos, x)  # 將 x_pos 映射到 x 的標籤

# 開啟格線顯示
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='red')  # 修正 which
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='black')  # 修正 which

plt.show()
```
