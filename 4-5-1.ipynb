```python
#建立Series物件-透過list建立
import pandas as pd
s=pd.Series([2,4,6,8,10]) #透過list建立pd名為s的Series物件
print(s)
s.value_counts(dropna=False) 
#針對數列成員的唯一值出現次數進行計數  
#參數dropna=False表示即使有缺失值（NaN），也會包含在計數結果中
print(s.value_counts(dropna=False))
```

```python
#建立Series物件2-透過ndarray建立
import pandas as pd
import numpy as np
data=np.array(['a','b','c','d'])
s=pd.Series(data)
print(s)
s=pd.Series(data,index=[3,5,7,9]) #自行建立索引(index)
print(s)
```

```python
#建立Series物件3-透過字典建立
import pandas as pd
import numpy as np
data={'a':0.,'b':1.,'c':2.}
s=pd.Series(data)
print(s)

s=pd.Series(data,index=['b','c','d','a'])
print(s)

s=pd.Series(5,index=[0,1,2,3])
print(s)
```

```python
#Series物件可透過索引存取
#負index也可用(但即將棄用)
import pandas as pd
s=pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
#顯示Series的前3個與最後3個成員
print(s[:3])
print(s[-3])
#Series物件可透過index存取
#提供的index不存在會報錯
print(s['a'])
print(s[['a','c','d']])
#注意錯誤訊息
print(s['f'])
```

```python
#建立DataFrame-透過random
import pandas as pd
import numpy as np
A=pd.DataFrame(np.random.rand(10,5))
A.head(3)
A.tail(2)
```

```python
#DataFrame基本特性

# 取得 DataFrame 的形狀，返回一個元組，顯示 (行數, 列數)
A.shape

# 顯示 DataFrame 的基本資訊，包括列名、數據類型、非空值計數和記憶體使用量
A.info

# 生成 DataFrame 的描述性統計資訊，如計數、均值、標準差、最小值、最大值和分位數（僅適用於數值列）
A.describe()

# 對 DataFrame 的每一列應用 pd.Series.value_counts，計算每個唯一值的出現頻次，返回一個新的 DataFrame
A.apply(pd.Series.value_counts)
```

```python
import pandas as pd
# 導入 Pandas 庫，用於數據處理和分析

import numpy as np
# 導入 NumPy 庫，用於生成隨機數據

df = pd.DataFrame(np.random.randn(4, 2), index=pd.date_range('2020-09-01', periods=4), columns=list('AB'))
# 創建一個 4 行 2 列的 DataFrame：
# - np.random.randn(4, 2) 生成 4x2 的隨機數陣列（標準正態分佈）
# - index=pd.date_range('2020-09-01', periods=4) 創建從 2020-09-01 開始的 4 天日期索引
# - columns=list('AB') 將列名設為 'A' 和 'B'

df.plot()
# 使用 Pandas 內建的繪圖功能，基於 Matplotlib，生成 DataFrame 的線圖：
# - 每個列（A 和 B）繪製為一條折線
# - x 軸為日期索引，y 軸為隨機數值
```

```python
#Pandas中類別式資料的處理
#直接設定類別式資料型態
import pandas as pd
s=pd.Series(['a','b','c','a'],dtype='category')
print(s)
```

```python
#Pandas中類別式資料的處理
#可設定類別順序和範圍
#Python以NaN取代不存在類別
import pandas as pd
cat=pd.Categorical(['a','b','c','a','b','c'])
cat2=pd.Categorical(['a','b','c','a','b','c','d'],['c','b','a'],ordered=True)
print(cat)
print(cat2)
```
